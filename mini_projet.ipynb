**Recherche des documents contenant le mot et le nombre d'occurnaces**

```python
def recherche_mot(mot, corpus):
    print("**Liste des documents contenant le mot " + mot + " avec leur index :**")
    documents = []
    nb_occurrences_par_document = []
    for i, doc in enumerate(corpus):
        if mot in doc:
            print(f"Document {i} : {doc}")
            nb_occurrences = doc.count(mot)
            documents.append((i, doc))
            nb_occurrences_par_document.append((i, nb_occurrences))
    print(nb_occurrences_par_document)
    return documents, nb_occurrences_par_document
```

**calcul du poids/tft/dft du mots dans chaque documents**

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from math import log

def poids_tf_idf_par_document(mot, corpus):
    resultats_rech, nb_occ_par_doc = recherche_mot(mot, corpus)
    poids_tf_idf_par_document = {}
    N = len(corpus)
    dft = len(resultats_rech)

    for i, (doc_index, doc) in enumerate(resultats_rech):
        tft_d = nb_occ_par_doc[i][1]
        poids = (1 + log(tft_d)) * log(N / dft)
        tfidf = (tft_d / dft) * log(N / dft)
        poids_tf_idf_par_document[doc_index] = {"poids": poids, "tfidf": tfidf, "document": doc}

    return poids_tf_idf_par_document

```

```python
def affichepoids(mot_recherche,corpus):
  resultats_poids_tf_idf = poids_tf_idf_par_document(mot_recherche, corpus)
  print("Poids TF-IDF par document pour le mot", mot_recherche)
  for doc_index, poids_tfidf in resultats_poids_tf_idf.items():
    print("Numéro du document:", doc_index)
    print("Poids:", poids_tfidf["poids"])
    print("TF-IDF:", poids_tfidf["tfidf"])
    print("Document:", poids_tfidf["document"])
```

**determination du  document plus pertinent**

```python
def document_plus_pertinent(mot, corpus):
    resultats_poids_tf_idf = poids_tf_idf_par_document(mot, corpus)
    document_pertinent = None
    max_score = float('-inf')

    for doc_index, poids_tfidf in resultats_poids_tf_idf.items():
        score = poids_tfidf["tfidf"]
        if score > max_score:
            max_score = score
            document_pertinent = (doc_index, poids_tfidf["document"])

    return document_pertinent
```

```python
def affichepertinent(mot_recherche,corpus):
  document_pertinent = document_plus_pertinent(mot_recherche, corpus)
  print("Document le plus pertinent pour le mot", mot_recherche, ":", document_pertinent)
```
